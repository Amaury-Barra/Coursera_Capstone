# Scrapping ALL the data of the URL using pd.read.html and vizualise the pd dataframes constituated

```python
dataframes = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
for df in dataframes:
    print(df)
```

# Choosing the relevant dataframe scrapped from the page
## vizualization was commented out

```python
toronto_data = dataframes[0]
#toronto_data
```

# Replace the Neighbourhood not assigned by their borough name
# Drop borough not assigned
### Vizualisation of the data set was commented out

```python
toronto_data.Neighbourhood.replace('Not assigned',toronto_data.Borough,inplace=True)
toronto_data = toronto_data[toronto_data.Borough != 'Neighbourhood']
#toronto_data
```

# Grouping by postcode and borough for unique idenfifier

```python
toronto_data_gp = toronto_data.groupby("Postcode").agg(lambda x:','.join(set(x)))
toronto_data_gp.reset_index(inplace=True)
toronto_data_gp
```

# Use shape method

```python
toronto_data_gp.shape
```

# Extracting a list of latitude and longitude

```python
lat=[] # initialize an empty list to store latitudes
lng=[] # initialize an empty list to store longitudes
for index, row in toronto_data_gp.iterrows(): #Loop through all the rows in the dataset
    lat_lng_coords = None
    postal_code=(row['Postcode']) # Strore in a variable to the postcode of each row (to pass it to the geocoder)
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Toronto, Ontario'.format(postal_code))
        lat_lng_coords = g.latlng
    lat.append(lat_lng_coords[0]) # append the retrieved lat from the geocoder to the lat list
    lng.append(lat_lng_coords[1]) # append the retrieved lng from the geocoder to the lng list
print(len(lat)) # check that the len of the list and df match
print(len(lng))
```

```python
toronto_data_gp['Latitude']= lat
toronto_data_gp['Longitude']= lng
toronto_data_gp
```
